# Moving Average Process Demo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define parameters of the MA process of order 2
np.random.seed(11)
mu = 4.0
theta1 = 0.5
theta2 = 0.3
stdEpsilon = 1.3

# define number of time steps
n = 25

# Generate a time series of white noise with mean zero and std stdEpsilon
epsilon_t = np.random.normal(0.0, stdEpsilon, n)

# plot the time series of white noise
plt.figure(figsize=(14,10))
plt.plot(epsilon_t)
plt.show()
```

```python
# create storage container for MA process
Z_t = np.zeros((n,1))
T = np.arange(0,n,1)

# Create the values of the noise process at timestep t-1 and t-2 to initiate the model
eps_tm1 = np.random.normal(0.0, stdEpsilon, 1)
eps_tm2 = np.random.normal(0.0, stdEpsilon, 1)

for t in T:
    if t == 0:
        Z_t[t] = mu + theta1 * eps_tm1 + theta2 * eps_tm2
    if t == 1:
        Z_t[t] = mu + theta1 * epsilon_t[t-1] + theta2 * eps_tm1
    if t > 1:
        Z_t[t] = mu + theta1 * epsilon_t[t-1] + theta2 * epsilon_t[t-2]
        

plt.figure(figsize=(14,10))
plt.plot(Z_t)
plt.show()
```
